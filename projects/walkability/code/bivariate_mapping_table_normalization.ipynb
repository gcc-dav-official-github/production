```python
def reclass_A(v):
    if v == 0:
        v = "Low"
    elif v == 1:
        v = "Middle"
    elif v == 2: 
        v = "High"
    return v
def reclass_N(v):
    if v == 0:
        v = "High"
    elif v == 1:
        v = "Middle"
    elif v == 2: 
        v = "Low"
    return v
```

```python
#a8 = 0 to 14 years
#a13 = 15 to 19 years
ct_join["akids"] =  ct_join["a8"]+ct_join["a13"]

#normalizing values
#assigning low, middle, high 
for col in ct_join.columns:
    if col.startswith('a'):
        mn = ct_join[col].min()
        mx = ct_join[col].max()
        #creates normalized values for each column with _norm, column is dropped
        #for example a1_norm to a2224_norm
        ct_join[col+"_norm"] = ct_join[col].apply(lambda x: (x-mn)/(mx-mn) )
        #creates class values by taking normalized values and applying 3 quantiles
        ct_join[col+"_class"]=ct_join[col+"_norm"].apply(py.Quantiles.make(k=3,rolling=True))
        #creates class values and reclass Low, Middle, High
        #for example a1_class to a2224_class
        ct_join[col+"_class"]=ct_join[col+"_class"].apply(lambda x: reclass_A(x) )
        ct_join.drop(col+"_norm",axis=1, inplace=True)
        #ct_join.drop(col,axis=1, inplace=True)
    if col.startswith('m') and "avg" in col:
        mn = ct_join[col].min()
        mx = ct_join[col].max()
        ct_join[col+"_norm"] = ct_join[col].apply(lambda x: (x-mn)/(mx-mn) )        
        ct_join[col+"_class"]=ct_join[col+"_norm"].apply(py.Quantiles.make(k=3,rolling=True))
        ct_join[col+"_class"]=ct_join[col+"_class"].apply(lambda x: reclass_N(x) )
        ct_join.drop(col+"_norm",axis=1, inplace=True)
        #ct_join.drop(col,axis=1, inplace=True)
    
        
               # ct_join[col+"_class"]=ct_join[col+"_class"].apply(lambda x: reclass(x) )   

ct_join.head(5)
```
