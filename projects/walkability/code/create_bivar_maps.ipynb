```python
#BiVariate plots for census tracts

#change project location and name
aprx = arcpy.mp.ArcGISProject(r"C:/Users/kmerrem/Data/Karl_Workspace/ProductsSupport/Sidewalk/Bivariate/Bruce/Bivariate.aprx")

#Bivariate.aprx is referencing classification_walkability
#update list after creating more relationship stats
#df = pd.read_csv('./bivariate_layer_list_details1.csv', encoding = 'latin-1')
#combined Bruce and my spreadsheet
df = pd.read_csv('./bivariate_layer_list_details4.csv', encoding = 'latin-1')

m = aprx.listMaps("Map2")[0]
layers = m.listLayers()
layout = aprx.listLayouts("BiVariate")[0]
lyr = layers[0]
# save symbology layer as layer then apply this symbology to other layers


# references layer where we keep symbology for the layer
# if you add additional columns to the classification_walkability layer then you will have to create a new layer file
# select the layer in ArcGIS Pro
# select Sharing
# select Save as Layer File

# Primary Symbology Unique Values
# HEX Values for symbology
# High - High 574249
# High - Low C85A5A
# High - Middle 985356
# Low - High 64ACBE
# Low - Low E8E8E8
# Low - Middle B0D5DF
# Middle - High 627F8C
# Middle - Low E4ACAC
# Middle - Middle AD9EA5

# Change symbology for <all other values> to 
# Color - no color
# Outline color - no color
# apply changes

#symbologylayer.lyrx has to be regenerated in arcgis pro if you add any new columns to the layer
symbolLyr = r"C:\Users\kmerrem\Data\Karl_Workspace\ProductsSupport\Sidewalk\Bivariate\Bruce\symbologylayer.lyrx"

sym = lyr.symbology

title = layout.listElements('TEXT_ELEMENT')[0]
legend_title = layout.listElements('TEXT_ELEMENT')[1]
legend_title_1 = layout.listElements('TEXT_ELEMENT')[2]
legend_title_1a = layout.listElements('TEXT_ELEMENT')[3]
sources = layout.listElements('TEXT_ELEMENT')[4]
variable2 = layout.listElements('TEXT_ELEMENT')[5]
mp = layout.listElements('MAPFRAME_ELEMENT')[0]


for ix, row in df.iterrows():
    render_var = row["render_var"]
    maptitle = row["title"]
    print (maptitle)
    legendtitle = row["legend_title"]
    legendtitle1 = row["legend_title_1"]
    legendtitle1a = row["legend_title_1a"]
    sourcetitle = row["sources"]
    variable2text = row["variable_2"]
    title.text = maptitle
    legend_title.text = legendtitle
    legend_title_1.text = legendtitle1
    legend_title_1a.text = legendtitle1a
    sources.text = sourcetitle
    variable2.text = variable2text   
    #colorbar = {"food":"Greens (Continuous)","health":"Orange-Red (Continuous)","publicservice":"Yellow-Orange-Brown (Continuous)","transit":"Purples (Continuous)"}
    #color = colorbar[theme]
    #sym.updateRenderer('UniqueValueRenderer')
    sym.renderer.fields = render_var
    #render_var = set_color(render_var)
    
    
    sym.renderer.classificationField = render_var
    sym.renderer.breakCount = 9
    #sym.renderer.colorRamp = aprx.listColorRamps(color)[0]
    lyr.symbology = sym
    # create symbology fields
    symbologyFields = [["VALUE_FIELD", render_var, render_var]]
    arcpy.ApplySymbologyFromLayer_management(lyr, symbolLyr, symbologyFields  )
  
    aprx.save()
    
#create output directory first    
    layout.exportToJPEG("./output_bivar/ct_" + render_var + ".jpg",600,jpeg_color_mode='24-BIT_TRUE_COLOR', jpeg_quality=75)
#    break
```
