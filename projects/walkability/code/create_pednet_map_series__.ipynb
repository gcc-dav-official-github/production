```python
aprx = arcpy.mp.ArcGISProject(r"C:/Users/Bwalker/Documents/ArcGIS/Projects/test_prod_mapping/test_prod_mapping.aprx")

df = pd.read_csv('./layer_list_details.csv', encoding = 'latin-1')

m = aprx.listMaps("CT")[0]
layers = m.listLayers()
layout = aprx.listLayouts("CT")[0]
lyr = layers[0]
sym = lyr.symbology

title = layout.listElements('TEXT_ELEMENT')[0]
sources = layout.listElements('TEXT_ELEMENT')[1]
legend = layout.listElements('LEGEND_ELEMENT')[0]
mp = layout.listElements('MAPFRAME_ELEMENT')[0]
for ix, row in df.iterrows():
    render_var = row["minutes_avg"]
#    maptitle = row["title"]
    maptitle = row["new_title"]
    print (maptitle)
    sourcetitle = row["source_note"]
    theme = row["theme"]
    title.text = maptitle
    sources.text = sourcetitle
    colorbar = {"food":"Greens (Continuous)","health":"Orange-Red (Continuous)","publicservice":"Yellow-Orange-Brown (Continuous)","transit":"Purples (Continuous)"}
    color = colorbar[theme]
    sym.updateRenderer('GraduatedColorsRenderer')
    sym.renderer.classificationField = render_var
    sym.renderer.breakCount = 6
    sym.renderer.colorRamp = aprx.listColorRamps(color)[0]
    #sym.renderer.symbol.outlineColor = {'CMYK' : [25, 50, 75, 25, 0]}
    lyr.symbology = sym
    aprx.save()
    layout.exportToJPEG("./output2/ct_" + render_var + ".jpg",600,jpeg_color_mode='24-BIT_TRUE_COLOR', jpeg_quality=75)
#    break
```
